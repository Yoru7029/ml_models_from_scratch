```python
from sklearn.datasets import fetch_california_housing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn import linear_model
import pandas as pd
import numpy as np

from my_models import reg
```

```python
housing = fetch_california_housing()

X = housing.data
y = housing.target
```

```python
X_scaled = StandardScaler().fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
X_train.shape, y_train.shape
```

```python
lr = linear_model.LinearRegression().fit(X_train, y_train)
my_lr = reg.LinearRegression().fit(X_train, y_train)
```

```python
print(lr.score(X_test, y_test))
print(my_lr.score(X_test, y_test))
```

```python
print("Sklearn model:")
print("     MSE:", mean_squared_error(y_test, lr.predict(X_test)))
print(f"    coefficients: {lr.coef_}\n\n  intercept: {lr.intercept_}\n")
print("My model:")
print("     MSE:", mean_squared_error(y_test, my_lr.predict(X_test)))
print(f"    coefficients: {my_lr.coef_}\n\n  intercept: {my_lr.intercept_}")
```

```python
my_lin_sgd = reg.SGDLinearRegression(tol=1e-3).fit(X_train, y_train)
```

```python
print("My SGD:")
print("     R2 score:", my_lin_sgd.score(X_test, y_test))
print("     MSE:", mean_squared_error(y_test, my_lin_sgd.predict(X_test)))
print("     intercept:", my_lin_sgd.intercept_)
print("     coefficent:", my_lin_sgd.coef_)
```
